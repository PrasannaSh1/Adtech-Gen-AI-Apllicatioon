```python
import pyspark as ps

from pyspark.sql import SparkSession

from pyspark.sql.functions import *

from pyspark.sql.window import Window

from pyspark.sql.types import *

spark = SparkSession.builder.appName("MyApp").getOrCreate()
```

```python
df = spark.read.csv("Adtech-Gen-AI-Apllicatioon\Data\LMS_DATA_1.csv",inferSchema=True, header=True)
```

```python
from pyspark.sql import functions as F 

#This imports the functions of pyspark.sql module in the form of an object, "F". "When" is one such function.

df9 = df.withColumn("More_Beer_Than_Wine_Servings", F.when(df.Number_of_Beer_Servings > df.Number_of_Wine_servings, 1).otherwise(0))

df9.show()
```
